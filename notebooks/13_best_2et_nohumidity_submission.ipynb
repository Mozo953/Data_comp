# Best 2ET Nohumidity Dirichlet Submission

Notebook wrapper for `scripts/train_best_2et_nohumidity_dirichlet.py`.


```python
from __future__ import annotations

import json
import subprocess
import sys
from pathlib import Path

import pandas as pd

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 180)
```

```python
command = [
    sys.executable,
    str(ROOT / "scripts" / "train_best_2et_nohumidity_dirichlet.py"),
    "--data-dir",
    "src/odor_competition/data",
]

run = subprocess.run(command, cwd=ROOT, capture_output=True, text=True, check=True)
raw_stdout = run.stdout
json_start = raw_stdout.index("{")
summary = json.loads(raw_stdout[json_start:])
print(raw_stdout)
```

```python
pd.DataFrame([
    {
        "model": summary["model"],
        "oof_weighted_wrmse": summary["simplex"]["oof_weighted_wrmse"],
        "submission_path": summary["artifacts"]["submission_path"],
        "summary_path": summary["artifacts"]["summary_path"],
    }
])
```
