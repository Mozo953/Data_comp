# YData Profiling - Gas Detection Dataset

Notebook reproductible pour profiler le dataset complet : `X_train`, `X_test` et `y_train`.

Les rapports HTML sont écrits dans `artifacts_ydata_profiling/`.

```python
from pathlib import Path
import sys
import pandas as pd
from ydata_profiling import ProfileReport

ROOT = Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))

from odor_competition.data import standardize_feature_columns, standardize_target_columns

DATA_DIR = ROOT / 'src' / 'odor_competition' / 'data'
OUT_DIR = ROOT / 'artifacts_ydata_profiling'
OUT_DIR.mkdir(parents=True, exist_ok=True)

DATA_DIR, OUT_DIR
```

```python
x_train = standardize_feature_columns(pd.read_csv(DATA_DIR / 'X_train.csv'))
x_test = standardize_feature_columns(pd.read_csv(DATA_DIR / 'X_test.csv'))
y_train = standardize_target_columns(pd.read_csv(DATA_DIR / 'y_train.csv'))

print('X_train:', x_train.shape)
print('X_test :', x_test.shape)
print('y_train:', y_train.shape)
display(x_train.head())
display(y_train.head())
```

```python
features_train = x_train.drop(columns=['ID'], errors='ignore').copy()
features_test = x_test.drop(columns=['ID'], errors='ignore').copy()
features_train['dataset_origin'] = 'train'
features_test['dataset_origin'] = 'test'
features_combined = pd.concat([features_train, features_test], axis=0, ignore_index=True)

features_combined.shape
```

```python
features_report = ProfileReport(
    features_combined,
    title='Gas detection dataset - X_train + X_test profiling',
    minimal=True,
    explorative=True,
    progress_bar=True,
)
features_report_path = OUT_DIR / 'gas_features_train_test_ydata_profile.html'
features_report.to_file(features_report_path)
features_report_path
```

```python
targets = y_train.drop(columns=['ID'], errors='ignore').copy()
targets_report = ProfileReport(
    targets,
    title='Gas detection dataset - y_train targets profiling',
    minimal=True,
    explorative=True,
    progress_bar=True,
)
targets_report_path = OUT_DIR / 'gas_targets_ydata_profile.html'
targets_report.to_file(targets_report_path)
targets_report_path
```

## Notes

- Les colonnes legacy sont standardisées en `Humidity`, `M12`, ..., `c01`, ... via `src/odor_competition/data.py`.
- `minimal=True` garde le rapport exploitable sans exploser le temps de génération.
- `dataset_origin` permet de voir le shift train/test dans le rapport features.
